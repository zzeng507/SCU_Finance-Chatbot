```python
import numpy as np

def df_to_text(df, name):

    df_clean = df.dropna(how="all", axis=0).dropna(how="all", axis=1)

    df_clean = df_clean.replace({np.nan: ""})

    lines = []
    lines.append(f"TABLE SOURCE: {name}")


    for _, row in df_clean.iterrows():
        parts = []
        for col, val in row.items():
            val = str(val).strip()
            if val != "":
                col_name = str(col).strip()
                parts.append(f"{col_name}: {val}")
        if parts:
            lines.append(" | ".join(parts))

    return "\n".join(lines)

text_finance = df_to_text_clean(df1)
text_re_minor = df_to_text(df2, "RE Minor Checksheet")
text_offerings = df_to_text(df3, "25_26 Tentative FNCE_RE Class Offerings")

print(text_finance[:600])
```

```python
import pdfplumber
import os

# Path to the two PDF files
pdf_checklist_path = os.path.join(folder, "FINAL_UPDATED_FINANCE_CHECKLIST.pdf")
pdf_flowchart_path = os.path.join(folder, "Finance Prereq Flow Chart (8.5 x 11 in) (2).pdf")

# Function to extract text from a PDF file
def pdf_to_text(path, name):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text() or ""   # Extract text or use empty string if None
            text += page_text + "\n"               # Append with newline for readability
    return f"PDF SOURCE: {name}\n" + text

# Extract text from both PDFs
text_pdf_checklist = pdf_to_text(pdf_checklist_path, "FINAL_UPDATED_FINANCE_CHECKLIST")
text_pdf_flowchart = pdf_to_text(pdf_flowchart_path, "Finance Prereq Flow Chart")

# Preview first 500 chars to confirm extraction
print(text_pdf_checklist[:500])
print("-" * 80)
print(text_pdf_flowchart[:500])
```

```python
import re
import numpy as np

# Function to clean text
def clean_text(text: str) -> str:
    # Replace tabs with spaces
    text = text.replace("\t", " ")
    # Collapse multiple whitespaces into one
    text = re.sub(r"\s+", " ", text)
    # Remove some irrelevant bullets or symbols
    text = text.replace("ÔÇß", " ").replace("‚óè", " ")
    return text.strip()

docs = []

# Add all documents (Excel + PDF) into a single list
for name, raw_text in [
    ("finance_excel_checklist", text_finance),
    ("re_minor_excel", text_re_minor),
    ("offerings_excel", text_offerings),
    ("finance_pdf_checklist", text_pdf_checklist),
    ("prereq_flowchart_pdf", text_pdf_flowchart),
]:
    docs.append({
        "source": name,
        "raw_text": raw_text,
        "clean_text": clean_text(raw_text)
    })

# Quick check
len(docs), [d["source"] for d in docs]
```

```python
# Function to chunk text into overlapping segments
def chunk_text(text, max_len=800, overlap=200):
    chunks = []
    start = 0
    while start < len(text):
        end = start + max_len
        chunk = text[start:end]
        chunks.append(chunk)
        start = end - overlap   # Overlap to avoid context loss
    return chunks

all_chunks = []

# Apply chunking to each document
for d in docs:
    chunks = chunk_text(d["clean_text"])
    for i, c in enumerate(chunks):
        all_chunks.append({
            "text": c,
            "source": d["source"],
            "chunk_id": i,
        })

# Preview one chunk
len(all_chunks), all_chunks[0]["source"], all_chunks[0]["text"][:300]
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

# Load embedding model (lightweight and good for RAG)
model = SentenceTransformer("all-MiniLM-L6-v2")

# Collect all chunk texts
texts = [c["text"] for c in all_chunks]

# Compute embeddings
embeddings = model.encode(texts, batch_size=32, show_progress_bar=True)
embeddings = np.array(embeddings).astype("float32")

embeddings.shape
```

```python
dim = embeddings.shape[1]

# Create a simple L2 FAISS index
index = faiss.IndexFlatL2(dim)

# Add embeddings to FAISS index
index.add(embeddings)

index.ntotal
```

```python
def search_docs(query, k=5):
    # Embed the user query
    q_emb = model.encode([query]).astype("float32")

    # FAISS search
    D, I = index.search(q_emb, k)

    results = []
    for idx, dist in zip(I[0], D[0]):
        item = all_chunks[idx]
        results.append({
            "source": item["source"],
            "chunk_id": item["chunk_id"],
            "text": item["text"],
            "score": float(dist),
        })
    return results
```

```python
# Build a RAG-style prompt using retrieved chunks
def build_rag_prompt(query, retrieved_chunks):
    # Combine top-k chunks into a single context block
    context = "\n\n".join([c["text"] for c in retrieved_chunks])

    prompt = f"""
You are an assistant that answers questions using ONLY the provided context.
If the answer is not in the context, say "I cannot find this information in the documents."

CONTEXT:
{context}

USER QUESTION:
{query}

FINAL ANSWER:
"""
    return prompt
```

```python
import os
import getpass
from openai import OpenAI

# Enter your OpenRouter API key
api_key = getpass.getpass("Enter your OpenRouter API key: ").strip()

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)
```

```python
# High-level RAG function
def ask_rag(question, k=5):
    # 1. Retrieve top-k relevant chunks
    retrieved = search_docs(question, k=k)

    # 2. Build LLM prompt with context
    prompt = build_rag_prompt(question, retrieved)

    # 3. Call LLM via OpenRouter
    answer = call_llm(prompt)

    return answer, retrieved
```

```python
# Function to call LLM using OpenRouter
def call_llm(prompt):
    response = client.chat.completions.create(
        model="mistralai/mistral-7b-instruct",   # or another OpenRouter model
        messages=[
            {"role": "system", "content": "You answer ONLY using the provided context."},
            {"role": "user", "content": prompt},
        ],
        temperature=0.0,
    )
    # IMPORTANT: use .content (attribute), NOT ["content"]
    return response.choices[0].message.content
```

```python
def search_docs(query, k=5):

    total = index.ntotal
    k = min(k, total)

    q_emb = model.encode([query]).astype("float32")
    D, I = index.search(q_emb, k)

    results = []
    for idx, dist in zip(I[0], D[0]):

        if idx < 0 or idx >= len(all_chunks):
            continue
        item = all_chunks[idx]
        results.append({
            "source": item["source"],
            "chunk_id": item["chunk_id"],
            "text": item["text"],
            "distance": float(dist),
        })
    return results
```

```python
question = "Based only on the Finance Prereq Flow Chart, after completing FNCE 124 and FNCE 125, which FNCE electives are shown as options? List all FNCE course numbers and names." # What are the prerequisites for FNCE 129? # What are the classes do I need to take to major in Finance?
answer, retrieved = ask_rag(question, k=5)

print("ANSWER:\n", answer)
print("="*80)
print("RETRIEVED CHUNKS USED:")
for c in retrieved:
    print(c["source"], "| chunk", c["chunk_id"])
```

```python
def get_fnce_chunks():
    return [
        c for c in all_chunks
        if ("fnce" in c["source"].lower()) or
           ("prereq_flowchart" in c["source"].lower())
    ]

fnce_chunks = get_fnce_chunks()
len(fnce_chunks)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

model = SentenceTransformer("all-MiniLM-L6-v2")

fnce_texts = [c["text"] for c in fnce_chunks]
fnce_emb = model.encode(fnce_texts).astype("float32")

fnce_index = faiss.IndexFlatL2(fnce_emb.shape[1])
fnce_index.add(fnce_emb)
```

```python
def ask_fnce(query, k=5):
    # search only FNCE chunks
    q_emb = model.encode([query]).astype("float32")
    D, I = fnce_index.search(q_emb, k)

    hits = []
    for idx, dist in zip(I[0], D[0]):
        if idx >= 0 and idx < len(fnce_chunks):
            hits.append(fnce_chunks[idx])

    # build context
    context = "\n".join([h["text"] for h in hits])

    prompt = f"""
You are ONLY allowed to answer using the FNCE Prereq Flow Chart data below.
Do NOT use Real Estate minor data or business core checklists.
If the chart does not explicitly list a course, do NOT mention it.



CONTEXT:
{context}

QUESTION:
{query}

ANSWER:
"""
    answer = ask_llm(prompt)
    return answer, hits
```

```python
answer, used = ask_fnce(
    "After completing FNCE 124 and FNCE 125, which FNCE electives are shown in the flow chart? List all course numbers."
)
print(answer)

print("\nCHUNKS USED:")
for h in used:
    print(h["source"], "|", h["chunk_id"])
```
